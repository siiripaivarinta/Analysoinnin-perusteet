```python
# Without loop structure
name = "Siiri"
guess = input ("Guess the name: ")
if guess == name:
    print("Congratulation!")
else:
    print("Guess again!")
```

```python
# With loop structure
name = "Siiri"
guess = ""
while guess != name:
    guess = input ("Guess the name: ")
    if guess == name:
        print("Congratulation!")
```

Subsetting your Data

```python
# Selecting a single column called Rating
subset_data = data["Rating"]
print(subset_data)
```

```python
# Selecting multiple columns
subset_data = data[["Rating", "Number of Reviews", "Release Date"]]
print(subset_data)
```

```python
# Getting ratings greater than 4.5
subset_data = data[data["Rating"] > 3.5]
print(subset_data)

# Subsetting based on text data
subset_data = data[data["Genres"] == "Shooter"]
print(subset_data)

# Subsetting with multiple conditions
subset_df1 = data["Rating"] > 3.5
subset_df2 = data["Genres"] > "Shooter"
subset_df3 = data[subset_df1 & subset_df2]
subset_df3.head() 
```

```python
# First lets create a function to define the rating ranges
def rating_category(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 3.5:
        return "Good"
    elif rating >= 2.5:
        return "Average"
    else:
        return "Poor"
    
# Now we can apply this function to the Rating column and create a new column called Rating Category
data["Rating Category"] = data["Rating"].apply(rating_category)
data.head()
```

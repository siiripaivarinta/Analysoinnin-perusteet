```python
# Exercise 1

# 1. Create list and assign furniture as values.
furniture = ["Sofa", "Table", "Chair", "Shelf"]

# 2. Print the whole list.
print(furniture)

# 3. Change the app so that only first two elements of the list are printed.
print(furniture[:2])

# 4. Change the app so that only “Sofa” is printed if it is found from the list. So, go through the whole list and check if the current element is “Sofa”.
for item in furniture:
    if item == "Sofa":
        print(item)
```

```python
# Exercise 2

import random
# 1. Create 5 element list. Name it thrownDiceNumbers.
thrownDiceNumbers = []

# 2. Pick randomly 5 numbers and assign them to the list. (google python randint())
for i in range(5):
    thrownDiceNumbers.append (random.randint (1, 6))

# 3. Print the whole list. 
print("Dice numbers:", thrownDiceNumbers)

# 4. Count the sum of the list values and print the result. 
total = sum(thrownDiceNumbers)
print("Sum of dice numbers:", total)

# 5. Print the highest value of the list. (Hint: Check built in function for list from the material)
highest = max(thrownDiceNumbers)
print("Highest dice number:", highest)
```
